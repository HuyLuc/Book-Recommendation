## MÔ TẢ TẬP DỮ LIỆU: 
- Tên data : books.csv
- File dữ liệu chứa 11123 thông tin của các cuốn sách.
- Mỗi dòng tương ứng với các thuộc tính sau của cuốn sách :
    - bookID (ID sách): Chứa ID duy nhất cho từng cuốn sách hoặc bộ sách.
    - title (Tiêu đề): Chứa tiêu đề của các cuốn sách.
    - authors (Tác giả): Chứa tên tác giả của từng cuốn sách cụ thể.
    - average_rating (Điểm trung bình): Là điểm trung bình của cuốn sách, được đánh giá bởi người dùng.
    - ISBN (Mã số ISBN 10): Là số ISBN(10), cung cấp thông tin về một cuốn sách như phiên bản và nhà xuất bản.
    - ISBN 13 (Mã số ISBN 13): Định dạng mới cho Mã số ISBN, được áp dụng từ năm 2007, có 13 chữ số.
    - language_code (Mã ngôn ngữ): Cho biết ngôn ngữ của cuốn sách.
    - Num_pages (Số trang): Chứa số trang của cuốn sách.
    - Ratings_count (Số lượng đánh giá): Chứa số lượng đánh giá mà người dùng đã đưa ra cho cuốn sách.
    - text_reviews_count (Số lượng đánh giá văn bản): Chứa số lượng bình luận hoặc đánh giá văn bản mà người dùng đã thể hiện cho cuốn sách.
    - publication_date: Ngày xuất bản của cuốn sách.
    - publisher: Nhà xuất bản của cuốn sách.

```python
# Phân tích năm xuất bản
df_books['publication_date'] = pd.to_datetime(df_books['publication_date'], errors='coerce')
df_books['publication_year'] = df_books['publication_date'].dt.year

# Loại bỏ các năm không hợp lệ
df_books_clean = df_books.dropna(subset=['publication_year'])
df_books_clean = df_books_clean[df_books_clean['publication_year'] > 1900]

# Tạo subplot với 2 biểu đồ
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Biểu đồ 1: Phân phối theo thập kỷ
df_books_clean['decade'] = (df_books_clean['publication_year'] // 10) * 10
decade_counts = df_books_clean['decade'].value_counts().sort_index()

decade_counts.plot(kind='bar', ax=ax1, color='lightblue')
ax1.set_title('Phân phối sách theo thập kỷ')
ax1.set_xlabel('Thập kỷ')
ax1.set_ylabel('Số lượng sách')
ax1.tick_params(axis='x', rotation=45)

# Biểu đồ 2: Xu hướng xuất bản theo năm (2000-2020)
recent_books = df_books_clean[df_books_clean['publication_year'] >= 2000]
yearly_counts = recent_books['publication_year'].value_counts().sort_index()

yearly_counts.plot(kind='line', ax=ax2, color='red', marker='o')
ax2.set_title('Xu hướng xuất bản sách (2000-2020)')
ax2.set_xlabel('Năm')
ax2.set_ylabel('Số lượng sách')

plt.tight_layout()
plt.show()

print(f"Năm xuất bản sớm nhất: {df_books_clean['publication_year'].min()}")
print(f"Năm xuất bản muộn nhất: {df_books_clean['publication_year'].max()}")
print(f"Thập kỷ có nhiều sách nhất: {decade_counts.idxmax()}s")
```

## Tác giả  Stephen King :
---

```python
def plot_author_chart(author_df):
    # Không thay đổi dữ liệu gốc
    df = author_df.copy()

    # Chuẩn hóa ngày và lấy năm xuất bản
    df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce')
    df['publication_year'] = df['publication_date'].dt.year
    df = df.dropna(subset=['publication_year'])
    df['publication_year'] = df['publication_year'].astype(int)

    # Thiết lập style
    plt.figure(figsize=(13, 6))

    # Thứ tự năm tăng dần
    order = sorted(df['publication_year'].unique())

    # Vẽ cột trung bình rating theo năm
    ax = sns.barplot(
        x='publication_year',
        y='average_rating',
        data=df,
        order=order,
        ci=None,
        palette=sns.color_palette('viridis', n_colors=len(order))
    )

    # Đường trung bình toàn bộ để tham chiếu
    ax.axhline(df['average_rating'].mean(), linestyle='--', linewidth=1)

    # Gắn nhãn giá trị lên mỗi cột
    for p in ax.patches:
        h = p.get_height()
        ax.text(
            p.get_x() + p.get_width()/2, h + 0.02,
            f'{h:.2f}',
            ha='center', va='bottom', fontsize=10, fontweight='bold'
        )

    # Tiêu đề và nhãn trục
    plt.title(f"Average rating of books over time, {df['authors'].iloc[0]}", fontsize=16, fontweight='bold')
    plt.xlabel('Publication year', fontsize=13, fontweight='bold')
    plt.ylabel('Average rating', fontsize=13, fontweight='bold')
    plt.xticks(rotation=30, ha='right')

    # Tinh gọn khung hình
    sns.despine(top=True, right=True)
    plt.tight_layout()
    plt.show()

# Ví dụ gọi hàm (không đổi tên biến)
author_df = df_books[(df_books['authors'] == authors[0]) & (df_books['language_code'] == 'eng')]
plot_author_chart(author_df)
```

### Nhận xét :
Biểu đồ thể hiện **điểm đánh giá trung bình (Average rating)** của các sách do **Stephen King** viết, phân bố theo từng năm xuất bản.

- **Giai đoạn điểm cao nổi bật**:
  - Năm **1996** đạt mức cao nhất **4.44**, cho thấy tác phẩm trong năm này được độc giả đánh giá rất tích cực.
  - Các năm **1980 (4.22)**, **1982 (4.16)** và **1995 (4.14)** cũng thuộc nhóm có điểm trung bình cao.

- **Giai đoạn điểm thấp**:
  - Năm **1992** ghi nhận mức thấp **3.51**.
  - Năm **2006** có điểm trung bình **3.50**, thấp nhất toàn bộ giai đoạn.
  - Một số năm khác như **1989 (3.87)**, **1990 (3.77)** và **2000 (3.75)** cũng có điểm dưới 4.

```python
def plot_author_chart_simple(author_df):
    # Sao chép dữ liệu để không ảnh hưởng bản gốc
    df = author_df.copy()

    # Chuyển thành năm
    df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce')
    df['publication_year'] = df['publication_date'].dt.year

    # Loại bỏ giá trị NaN
    df = df.dropna(subset=['publication_year'])
    df['publication_year'] = df['publication_year'].astype(int)

    # Vẽ biểu đồ đơn giản
    plt.figure(figsize=(10, 5))
    order = sorted(df['publication_year'].unique())

    ax = sns.barplot(
        x='publication_year',
        y='average_rating',
        data=df,
        order=order,
        ci=None
    )

    # Thêm số trên đầu cột
    for p in ax.patches:
        height = p.get_height()
        ax.text(
            p.get_x() + p.get_width()/2,  # vị trí ngang
            height + 0.02,               # vị trí dọc
            f'{height:.2f}',              # giá trị hiển thị
            ha='center', va='bottom', fontsize=9
        )

    plt.title(f"Average rating of books over time, {df['authors'].iloc[0]}")
    plt.xlabel('Publication year')
    plt.ylabel('Average rating')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()

# Gọi hàm
author_df = df_books[(df_books['authors'] == authors[1]) & (df_books['language_code'] == 'eng')]
plot_author_chart_simple(author_df)
```

### Nhận xét : 
- Agatha Christie được xem là một huyền thoại suốt nhiều năm qua. Hầu hết các tác phẩm của bà đều đạt được đánh giá cao, vẫn duy trì ở mức rất cao.

- **Điểm cao nhất**: Năm 1990 đạt 4.39, vượt trội so với các năm khác.  
- **Các năm nổi bật khác**: 1997 (4.32) và 1998 (4.31) giữ mức đánh giá rất cao.  
- **Điểm thấp**: Năm 2008 thấp nhất với 3.79.  
- **Giai đoạn 1999–2010**: Dao động chủ yếu trong khoảng 3.8–4.2, ít biến động mạnh.  
- **Xu hướng chung**:  
  - 1990–1998: Điểm trung bình cao và ổn định.  
  - Sau 1998: Có xu hướng giảm nhẹ và duy trì ở mức trung bình–khá.  

## Tác giả Dan Brown :
---

```python
def plot_author_chart(author_df):
    # Làm việc trên bản sao để tránh thay đổi dữ liệu gốc
    df = author_df.copy()

    # Chuyển sang datetime và lấy năm
    df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce')
    df['publication_year'] = df['publication_date'].dt.year

    # Loại bỏ năm NaN
    df = df.dropna(subset=['publication_year'])
    df['publication_year'] = df['publication_year'].astype(int)

    # Tạo plot
    plt.figure(figsize=(8, 8))
    order = sorted(df['publication_year'].unique())

    ax = sns.barplot(
        x='publication_year',
        y='average_rating',
        data=df,
        order=order,
        palette='deep',
        ci=None
    )

    # Thêm số trên đầu mỗi cột
    for p in ax.patches:
        h = p.get_height()
        ax.text(
            p.get_x() + p.get_width()/2,
            h + 0.02,
            f'{h:.2f}',
            ha='center',
            va='bottom',
            fontsize=9
        )

    # Tiêu đề và nhãn
    try:
        author_name = author_df['authors'].iloc[2]
    except IndexError:
        author_name = author_df['authors'].iloc[0]
    plt.title(f"Average rating of books over time, {author_name}")
    plt.xlabel('Publication year')
    plt.ylabel('Average rating')
    plt.xticks(rotation=30)
    sns.despine(top=True, right=True)
    plt.tight_layout()
    plt.show()

# Gọi hàm
author_df = df_books[(df_books['authors'] == authors[2]) & (df_books['language_code'] == 'eng')]
plot_author_chart(author_df)
```

```python
# Hàm phên chia average_rating
def segregation(data):
    values = []
    for val in data.average_rating:
        if val>=0 and val<=1:
            values.append("Between 0 and 1")
        elif val>1 and val<=2:
            values.append("Between 1 and 2")
        elif val>2 and val<=3:
            values.append("Between 2 and 3")
        elif val>3 and val<=4:
            values.append("Between 3 and 4")
        elif val>4 and val<=5:
            values.append("Between 4 and 5")
        else:
            values.append("NaN")
    print(len(values))
    return values
```

Phân cụm KMeans là một kiểu học không giám sát, nhóm dữ liệu không được gắn nhãn. Mục tiêu là tìm các nhóm trong dữ liệu có điểm tương đồng với nhau.

Tìm kiếm  mối quan hệ giữa tổng số lượng đánh giá (ratings_count) với giá trị đánh giá trung bình (average_rating)

```python
trial = df_books[['average_rating', 'ratings_count']]
data = np.asarray([np.asarray(trial['average_rating']), np.asarray(trial['ratings_count'])]).T

```

Elbow curve (đôi khi còn được gọi là phương pháp "Elbow Method") thường được sử dụng trong quá trình lựa chọn số lượng cụm (clusters) phù hợp khi thực hiện thuật toán phân cụm (clustering), đặc biệt là trong phương pháp K-means clustering. Mục tiêu chính là xác định số lượng cụm tối ưu. Ta sẽ sử dụng phương pháp Elbow Curve để tìm cụm dữ liệu tốt nhất trong tập dữ liệu này


```python
X = data
distortions = []
for k in range(2,30):
    k_means = KMeans(n_clusters = k)
    k_means.fit(X)
    distortions.append(k_means.inertia_)

fig = plt.figure(figsize=(15,10))
plt.plot(range(2,30), distortions, 'bx-')
plt.title("Elbow Curve")
```

Quan sát biểu đồ ta nhận thấy rằng ở điểm số 5 là nơi mà sự giảm đáng kể trong tổng bình phương khoảng cách giảm dần và trở nên đều đặn. Chọn giá trị k=5.

```python
#k=5 nên ta lấy 5 cụm
centroids, _ = kmeans(data, 5)

#Gán từng mẫu vào một cụm
#Lượng tử hóa vecto

idx, _ = vq(data, centroids)
```

## Công cụ đề xuất

Sau khi xem phân cụm, chúng tôi có thể suy ra rằng có thể có một số đề xuất có thể xảy ra với mối quan hệ giữa Xếp hạng trung bình và Số lượng xếp hạng. Lấy Xếp hạng_Phân phối (Xu hướng phân loại tự tạo), hệ thống đề xuất hoạt động với thuật toán của K Hàng xóm gần nhất. Dựa trên cuốn sách do người dùng nhập, những cuốn sách gần nhất với cuốn sách đó sẽ được phân loại là những cuốn sách mà người dùng có thể thích. KNN được sử dụng cho cả vấn đề phân loại và hồi quy. Trong các bài toán phân loại để dự đoán nhãn của một phiên bản, trước tiên chúng tôi tìm k phiên bản gần nhất với phiên bản đã cho dựa trên thước đo khoảng cách và dựa trên sơ đồ bỏ phiếu đa số hoặc bỏ phiếu đa số có trọng số (những người lân cận gần hơn sẽ có trọng số cao hơn), chúng tôi dự đoán nhãn.

Trong bối cảnh như thế này, quá trình học tập không giám sát diễn ra với những hàng xóm tương tự được đề xuất. Đối với danh sách nhất định, nếu tôi hỏi đề xuất về "The Catcher in the Rye", năm cuốn sách liên quan đến nó sẽ xuất hiện

Tạo bảng tính năng sách, dựa trên Phân phối xếp hạng, phân loại sách theo thang xếp hạng, chẳng hạn như:
* Between 0 and 1
* Between 1 and 2
* Between 2 and 3
* Between 3 and 4
* Between 4 and 5

Nhìn rộng ra, các đề xuất sau đó sẽ xem xét xếp hạng trung bình và xếp hạng cout cho truy vấn đã nhập.

Tạo các hàm cụ thể để giúp tìm tên sách:

* Lấy chỉ mục từ Tiêu đề
* Nhận ID từ một phần tên (Vì không phải ai cũng có thể nhớ hết tên)
* In những cuốn sách tương tự từ tập dữ liệu tính năng. (Điều này sử dụng số liệu Chỉ số từ những người hàng xóm gần nhất để chọn sách.)

```python
#lấy chỉ số index từ đầu vào tên sách
def get_index_from_name(name):
    return df_books[df_books["title"]==name].index.tolist()[0]

all_books_names = list(df_books.title.values)

#lấy id từ một phần tên sách
def get_id_from_partial_name(partial):
    for name in all_books_names:
        if partial in name:
            print(name,all_books_names.index(name))

#đưa ra những cuốn sách tương tự             
def print_similar_books(query=None,id=None):
    if id:
        for id in indices[id][1:]:
            print(df_books.iloc[id]["title"])
    if query:
        found_id = get_index_from_name(query)
        for id in indices[found_id][1:]:
            print(df_books.iloc[id]["title"])
```

```python
# Đánh giá mô hình gợi ý sách (Unsupervised Learning)
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

print("=== ĐÁNH GIÁ MÔ HÌNH GỢI Ý SÁCH (UNSUPERVISED LEARNING) ===\n")

# 1. Đánh giá chất lượng phân cụm (Clustering Quality)
print("1. ĐÁNH GIÁ CHẤT LƯỢNG PHÂN CỤM:")
print("-" * 40)

# Tính Silhouette Score cho việc phân cụm
if 'data' in locals() and 'idx' in locals():
    silhouette_avg = silhouette_score(data, idx)
    print(f"Silhouette Score: {silhouette_avg:.3f}")
    print("   (Càng gần 1 thì chất lượng phân cụm càng tốt)")
else:
    print("Chưa có dữ liệu phân cụm để đánh giá")

print("\n2. ĐÁNH GIÁ ĐỘ ĐA DẠNG GỢI Ý:")
print("-" * 40)

# Đánh giá độ đa dạng của các gợi ý
def evaluate_diversity():
    # Lấy mẫu một số sách để test
    sample_books = df_books.sample(n=5)
    
    diversity_scores = []
    for idx, book in sample_books.iterrows():
        try:
            book_index = get_index_from_name(book['title'])
            recommended_indices = indices[book_index][1:6]  # Top 5 gợi ý
            
            # Tính độ đa dạng dựa trên tác giả và thể loại
            recommended_books = df_books.iloc[recommended_indices]
            unique_authors = len(recommended_books['authors'].unique())
            
            # Độ đa dạng = số tác giả unique / tổng số gợi ý
            diversity = unique_authors / len(recommended_indices)
            diversity_scores.append(diversity)
            
            print(f"Sách: {book['title'][:50]}...")
            print(f"   Độ đa dạng tác giả: {diversity:.2f}")
            
        except:
            continue
    
    if diversity_scores:
        avg_diversity = np.mean(diversity_scores)
        print(f"\nĐộ đa dạng trung bình: {avg_diversity:.3f}")
        print("   (Càng cao thì gợi ý càng đa dạng)")
    
    return diversity_scores

diversity_scores = evaluate_diversity()

print("\n3. ĐÁNH GIÁ ĐỘ TƯƠNG ĐỒNG GỢI Ý:")
print("-" * 40)

# Đánh giá độ tương đồng giữa sách gốc và sách được gợi ý
def evaluate_similarity():
    sample_books = df_books.sample(n=3)
    
    for idx, book in sample_books.iterrows():
        try:
            book_index = get_index_from_name(book['title'])
            recommended_indices = indices[book_index][1:4]  # Top 3 gợi ý
            
            print(f"\nSách gốc: {book['title']}")
            print(f"Rating: {book['average_rating']:.2f}, Đánh giá: {book['ratings_count']}")
            print("Gợi ý:")
            
            for i, rec_idx in enumerate(recommended_indices, 1):
                rec_book = df_books.iloc[rec_idx]
                rating_diff = abs(book['average_rating'] - rec_book['average_rating'])
                print(f"   {i}. {rec_book['title'][:40]}...")
                print(f"      Rating: {rec_book['average_rating']:.2f} (chênh lệch: {rating_diff:.2f})")
                
        except:
            continue

evaluate_similarity()

print("\n4. PHÂN TÍCH HIỆU QUẢ MÔ HÌNH:")
print("-" * 40)
print("✓ Mô hình sử dụng KMeans clustering và KNN recommendation")
print("✓ Dữ liệu được chuẩn hóa bằng MinMaxScaler")
print("✓ Sử dụng ball_tree algorithm cho tìm kiếm nearest neighbors")
```
